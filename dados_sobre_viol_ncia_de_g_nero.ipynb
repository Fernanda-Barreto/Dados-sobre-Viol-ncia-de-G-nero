# Segunda representação gráfica


2. Levando em consideração a descrição do dataset fornecido, foi observado que as entradas na coluna 'Value' representadas por NaN indicam que tais valores correspondem às pessoas que não concordaram com a pergunta realizada.



```python
# criando um Df filtrando os valores de Value
concorda_df = df.dropna(subset=['Value'])
nao_concorda_df = df[df['Value'].isna()]
# Contar o número de pessoas que concordam e não concordam
concorda_count = len(concorda_df)
nao_concorda_count = len(nao_concorda_df)
```

```python
# Dados para o primeiro gráfico
compare_df = pd.DataFrame({'Opinião': ['Concorda', 'Não Concorda '], 'Count': [concorda_count, nao_concorda_count]})

#layout
color_discrete_map = {'Pergunta A': 'blue', 'Pergunta B': 'green', 'Pergunta C': 'orange'}

#criação do grafico
pie_fig = px.pie(compare_df, names='Opinião', values='Count',
                 title='Distribuição entre Pessoas que Concordam e Não Concordam com algum tipo de agressão',
                 color='Opinião',
                 color_discrete_sequence=px.colors.sequential.Blues_r)  # Usando a mesma paleta de cores
pie_fig.update_layout(template='plotly_dark')

pie_fig.show()
py.iplot(pie_fig)
#pyo.plot(pie_fig)
#pie_fig.write_html('grafico3.html')

# Baixe o arquivo
#files.download('grafico3.html')
```

# Nona representação gráfica

1.
Com a avaliação dos dois últimos gráficos, foi elaborado um gráfico de comparação para uma melhor visualização dos dados

```python
#primeiro conjunto de dados
contagem_anos_Nconcordancia = contagem_anos_Nconcordancia.append(pd.Series([0], index=[pd.to_datetime('2004-01-01')]))
df_contagem_anos_Nconcordancia = pd.DataFrame({'Survey Year': contagem_anos_Nconcordancia.index, 'Quantity': contagem_anos_Nconcordancia.values})
df_contagem_anos_Nconcordancia['Survey Year'] = pd.to_datetime(df_contagem_anos_Nconcordancia['Survey Year'])
df_contagem_anos_Nconcordancia = df_contagem_anos_Nconcordancia.sort_values(by='Survey Year')

#segundo conjunto de dados
contagem_anos_concordancia = contagem_anos_concordancia.append(pd.Series([0], index=[pd.to_datetime('2004-01-01')]))
df_contagem_anos_concordancia = pd.DataFrame({'Survey Year': contagem_anos_concordancia.index, 'Quantity': contagem_anos_concordancia.values})
df_contagem_anos_concordancia['Survey Year'] = pd.to_datetime(df_contagem_anos_concordancia['Survey Year'])
df_contagem_anos_concordancia = df_contagem_anos_concordancia.sort_values(by='Survey Year')

#juntando os dois
df_combined = pd.concat([df_contagem_anos_Nconcordancia, df_contagem_anos_concordancia], keys=['Disagrees', 'Agrees'])

#layout
fig_combined = px.line(df_combined, x='Survey Year', y='Quantity', title='Trends in the acceptance and disapproval of violent behaviors',
                       color=df_combined.index.get_level_values(0))

fig_combined.update_layout(template='plotly_dark')

fig_combined.update_layout(
    annotations=[
        dict(
            text='• An analysis of the average agreement concerning different justifications for aggression',
            showarrow=False,
            x=-0.02, y=1.1,
            xref="paper", yref="paper",
            align="center",
            font=dict(size=9)
        )
    ])

# Exibindo o gráfico combinado
fig_combined.show()

py.iplot(fig_combined)
#pyo.plot(fig_combined)
#fig_combined.write_html('grafico9.html')

# Baixe o arquivo
#files.download('grafico9.html')
```

# Decima representação gráfica

1.tipos de justificativas por genero e faixa etaria


```python
df_idades = df_filtered[df_filtered['Demographics Response'].isin(['15-24'])]
df_idades
```

```python
valores = df_idades['Question'].unique()
valores
```

```python
import plotly.express as px

# Filtrar apenas as faixas etárias '15-24', '25-34' e '35-49'
filtered_df = df_filtered[df_filtered['Demographics Response'].isin(['15-24', '25-34', '35-49'])]

pivot_df = filtered_df.pivot_table(index=['Question', 'Demographics Response'], columns='Gender', values='Value', aggfunc='max').reset_index()

max_values_df = pivot_df.melt(id_vars=['Question', 'Demographics Response'], var_name='Gender', value_name='Value')

#layout
fig_demographics_filtered_max_values = px.bar(max_values_df,

                                               x="Demographics Response",
                                               y="Value",
                                               color="Question",
                                               facet_col="Gender",
                                               barmode="stack",
                                               color_discrete_sequence=px.colors.sequential.Blues_r)

fig_demographics_filtered_max_values.update_layout(title=" ", title_x=0.5,title_y=0.99 )

fig_demographics_filtered_max_values.update_layout(template='plotly_dark')


fig_demographics_filtered_max_values.update_layout(yaxis=dict(showticklabels=False))

fig_demographics_filtered_max_values.show()

# Baixe o arquivo
#fig_demographics_filtered_max_values.write_html('grafico10.html')
#files.download('grafico10.html')
#salvando plotly
py.iplot(fig_demographics_filtered_max_values)
#pyo.plot(fig_demographics_filtered_max_values)
```
